```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns # data visualization
import random
from matplotlib import pyplot as plt

import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_rows', None)
```

## Reading the Files

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
```

```python
test_df.drop(columns=['PassengerId', 'Name', 'Ticket','Cabin'], inplace=True)
```

```python
test_df.info()
```

```python
# filling the missing age data with random values
test_df.Age.fillna(int(random.normalvariate(age_mean, age_std)), inplace=True)
test_df.Fare.fillna(test_df.Fare.median(), inplace=True)
```

```python
test_df = test_df[['Age', 'Pclass', 'Sex', 'SibSp', 'Parch', 'Fare',
       'Embarked']]
```

```python
test_df.head()
```

```python
test_df['Sex'] = test_df['Sex'].map({'male' : 0, 'female' : 1})
test_df['Embarked'] = test_df['Embarked'].map({'S' : 0, 'C' : 1, 'Q' : 2})
```

```python
train_df
```

```python
x = train_df[test_df.columns]
y = train_df['Survived']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

SEED = 5
np.random.seed(SEED)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, 
                                                    stratify=y, random_state=SEED)
model = LinearSVC()
model.fit(x_train, y_train)

predict = model.predict(x_test)

acc_score = accuracy_score(y_test, predict)
print('Score: '+str(round(acc_score*100, 2))+'%')
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

SEED = 5
np.random.seed(SEED)
x_train_raw, x_test_raw, y_train, y_test = train_test_split(x, y, test_size=0.25, 
                                                    stratify=y, random_state=SEED)

scaler = StandardScaler()
scaler.fit(x_train_raw)
x_train = scaler.transform(x_train_raw)
x_test = scaler.transform(x_test_raw)

model = LinearSVC()
model.fit(x_train, y_train)
predict = model.predict(x_test)

acc_score = accuracy_score(y_test, predict)
print('Score: '+str(round(acc_score*100, 2))+'%')
```

```python
predict = model.predict(test_df)
```

```python
results = test_df.copy()
results['Survived_LinearSVC'] = predict
```
